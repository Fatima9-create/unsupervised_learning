```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data_array = scaler.fit_transform(numeric_df) 
df_scaled_sklearn = pd.DataFrame(scaled_data_array, columns=numeric_df.columns)
```

```python
#inertia_ → WCSS (Within-Cluster Sum of Squares) deməkdir.
wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(df_scaled_sklearn)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(np.arange(1,11),wcss)
```

```python
#6)KPrototip usulu
```

```python
random_sample_df.isnull().sum()
```

```python
from kmodes.kprototypes import KPrototypes

df_clean = random_sample_df.drop(['Customer ID', 'Purchase History'], axis=1)

# Null datalari doldururuq
df_clean = df_clean.fillna(0)

#  Kateqorik data hardadisa onu tapiriq
cat_columns_idx = [df_clean.columns.get_loc(col) for col in df_clean.select_dtypes('object').columns]
```

```python
kproto = KPrototypes(n_clusters=3, init='Cao', verbose=1, random_state=42)
clusters = kproto.fit_predict(df_clean.values, categorical=cat_columns_idx)
random_sample_df['Klaster'] = clusters

random_sample_df[['Customer ID', 'Klaster']].head()
```

```python
#7)DBSCAN ile outlierleri tap
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

numeric_cols = ['Age', 'Income Level', 'Coverage Amount', 'Premium Amount']
df_numeric = random_sample_df[numeric_cols].dropna()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_numeric)


```

```python
dbscan = DBSCAN(eps=0.5, min_samples=5)
outliers = dbscan.fit_predict(X_scaled)

random_sample_df['Outlier_Status'] = outliers

total_outliers = (random_sample_df['Outlier_Status'] == -1).sum()
print(f"Total outliers: {total_outliers}")
print(random_sample_df[random_sample_df['Outlier_Status'] == -1].head())
```

```python
#5Agglomerative clustering (qiymetlendirme)
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

cols = ['Age', 'Income Level', 'Coverage Amount', 'Premium Amount']
df_numeric = random_sample_df[cols].dropna()

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_numeric)

agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
clusters = agg.fit_predict(X_scaled)

df_numeric['Cluster_Label'] = clusters


plt.figure(figsize=(12, 7))
linkage_matrix = linkage(X_scaled, method='ward')
dendrogram(linkage_matrix, truncate_mode='level', p=5)
plt.show()

print("Klasterləş")

```

```python
#8)Anomaly detection
```

```python
from sklearn.datasets import make_blobs
```

```python
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

numeric_features = ['Age', 'Income Level', 'Coverage Amount', 'Premium Amount']
df_clean = random_sample_df.dropna(subset=numeric_features).copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_clean[numeric_features])
```

```python
#isolation foresti tetbiq eliyek
iso_forest = IsolationForest(contamination=0.02, random_state=42)
anomalies = iso_forest.fit_predict(X_scaled)

df_clean['Anomaly_Score'] = anomalies
outliers = df_clean[df_clean['Anomaly_Score'] == -1]

print(f"Anomaliya sayı: {len(outliers)}")
```

```python
plt.figure(figsize=(12, 7))
sns.set_style("whitegrid")

sns.scatterplot(data=df_clean, x='Income Level',y='Coverage Amount', hue='Anomaly_Score', palette={1: 'blue', -1: 'red'}, alpha=0.6,edgecolor=None)
plt.show()
```

```python
#9DImensionality reduction.
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#  Numerical sutunlari secek(tarixnen id ni silek)
num_cols = random_sample_df.select_dtypes(include=['number']).drop(columns=['Klaster'], errors='ignore')

# NaN -lari median ilə doldururuq
num_cols_filled = num_cols.fillna(num_cols.median())


scaler = StandardScaler()
scaled_data = scaler.fit_transform(num_cols_filled)
```

```python
pca = PCA(n_components=2)
pca_results = pca.fit_transform(scaled_data)


pca_df = pd.DataFrame(data=pca_results, columns=['PC1', 'PC2'])
pca_df['Klaster'] = random_sample_df['Klaster'].values

# Scatter plot ile gosterek
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PC1', y='PC2', hue='Klaster', data=pca_df, palette='viridis', s=60, alpha=0.8)
plt.show()
```
